```python
import pandas as pd
import wikipedia

# Set the language for Wikipedia
wikipedia.set_lang("en")

# Assuming df is your DataFrame and 'Toponyms - Book' is the column of interest
# Example: df = pd.read_excel('your_file.xlsx')

# Initialize new columns
df['Title'] = None
df['URL'] = None
df['Image URL'] = None

# Iterate through each row and update the new columns
for index, row in df.iterrows():
    try:
        # Use Wikipedia's search to get suggestions and take the first result
        search_results = wikipedia.search(row['Toponyms - Book'], results=1)
        if search_results:
            page = wikipedia.page(search_results[0])
            df.at[index, 'Title'] = page.title
            df.at[index, 'URL'] = page.url
            df.at[index, 'Image URL'] = page.images[0] if len(page.images) > 0 else None
    except Exception as e:
        print(f"Error processing {row['Toponyms - Book']}: {e}")

# Now df has three new columns: 'Title', 'URL', and 'Image URL'
display(df)
```

```python
import pandas as pd
import os
import json

# Directory containing the chapter JSON files
directory = 'Data/json'

# Initialize a list to store DataFrames for each chapter
chapters_data = []

# Assuming you have a known number of chapters
num_chapters = 21

# Process each chapter
for chapter in range(1, num_chapters + 1):
    file_path = os.path.join(directory, f'chapter{chapter}.json')

    # Check if file exists
    if os.path.exists(file_path):
        try:
            # Read the JSON file
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

            # Convert to DataFrame
            df = pd.DataFrame(data)
            df['Chapter'] = chapter  # Add a column for the chapter number

            # Append the DataFrame to the list
            chapters_data.append(df)
        except json.JSONDecodeError:
            print(f"Error reading {file_path}: File may be empty or not valid JSON.")
        except Exception as e:
            print(f"Error processing {file_path}: {e}")

# Concatenate all DataFrames into a single DataFrame
consolidated_df = pd.concat(chapters_data, ignore_index=True) if chapters_data else pd.DataFrame()

# Output Excel file
output_file = 'consolidated_chapters.xlsx'

# Save the DataFrame to an Excel file
consolidated_df.to_excel(output_file, index=False)

print(f'All chapters have been consolidated into {output_file}')
```

```python
import pandas as pd
import requests
from PIL import Image
import io
import math

# Load the Excel file
df = pd.read_excel('data/final.xlsx')

# Filter for rows where 'img' column is 1 and drop duplicates
unique_urls = df[df['img'] == 1]['Image'].drop_duplicates().tolist()

# Download images
images = []
for url in unique_urls:
    try:
        response = requests.get(url)
        image = Image.open(io.BytesIO(response.content))
        images.append(image)
    except Exception as e:
        print(f"Error downloading {url} {e}")

# Check if there are any images to process
if not images:
    print("No valid images to create a collage.")
else:
    # Determine the collage layout
    num_images = len(images)
    collage_width = int(math.sqrt(num_images))
    collage_height = math.ceil(num_images / collage_width)

    # Create a new blank image for the collage
    width, height = max(img.size for img in images)
    collage = Image.new('RGB', (width * collage_width, height * collage_height))

    # Paste images into the collage
    for i, image in enumerate(images):
        x = i % collage_width * width
        y = i // collage_width * height
        collage.paste(image, (x, y))

    # Save the collage
    collage.save('collage_v3.jpg')
    print("Collage created and saved as 'collage.jpg'")
```

```python
import pandas as pd
import requests
from PIL import Image
import io
import math
import re

def get_original_image_url(url):
    # Pattern to find the '/thumb/' part and resolution specification in Wikimedia URLs
    pattern = re.compile(r'/thumb(/.+/[\w-]+)\.\w+/\d+px-[\w-]+\.\w+')
    match = pattern.search(url)
    if match:
        # Construct the URL for the original image
        return f'https://upload.wikimedia.org/wikipedia/commons{match.group(1)}'
    return url  # Return the original URL if no match

# Load the Excel file
df = pd.read_excel('data/final.xlsx')

# Filter for rows where 'img' column is 1 and drop duplicates
unique_urls = df[df['img'] == 1]['Image'].drop_duplicates().tolist()

# Download images
images = []
for url in unique_urls:
    try:
        original_url = get_original_image_url(url)
        response = requests.get(original_url)
        image = Image.open(io.BytesIO(response.content))
        images.append(image)
    except Exception as e:
        print(f"Error downloading or processing {url}: {e}")

# Check if there are any images to process
if not images:
    print("No valid images to create a collage.")
else:
    # Determine the collage layout
    num_images = len(images)
    collage_width = int(math.sqrt(num_images))
    collage_height = math.ceil(num_images / collage_width)

    # Create a new blank image for the collage
    width, height = max(img.size for img in images)
    collage = Image.new('RGB', (width * collage_width, height * collage_height))

    # Paste images into the collage
    for i, image in enumerate(images):
        x = i % collage_width * width
        y = i // collage_width * height
        collage.paste(image, (x, y))

    # Save the collage
    collage.save('collage_v2.jpg')
```
